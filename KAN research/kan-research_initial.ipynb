# Introduction: Understanding KAN

Kolmogorov-Arnold Networks (KANs) are an innovative type of neural network architecture that fundamentally differs from the conventional neural network models like Multi-Layer Perceptrons (MLPs). Inspired by the Kolmogorov-Arnold representation theorem, KANs offer significant improvements in terms of interpretability, accuracy, and parameter efficiency compared to traditional models. The introduction of KANs heralds a shift in how we understand neural network architectures, providing new paradigms in the way models are trained, optimized, and evaluated.

## Evolution of Neural Networks and the Need for KANs

Neural networks have undergone significant transformations since their inception. The earliest models, such as perceptrons, paved the way for subsequent architectures like convolutional neural networks (CNNs) and recurrent neural networks (RNNs). These models evolved in complexity and capability, culminating in highly sophisticated architectures like deep neural networks (DNNs) and transformers.

However, despite their success in various fields such as computer vision, natural language processing, and reinforcement learning, traditional neural networks face several challenges:
1. **Interpretability**: Understanding the internal decision-making process is challenging due to the "black-box" nature of these models. This lack of transparency raises issues related to bias, trustworthiness, and safety in critical applications like healthcare and autonomous systems.
2. **Efficiency**: High parameter count and computational requirements make training, maintaining, and deploying these networks costly and sometimes impractical.
3. **Overfitting**: Due to their high flexibility, neural networks can easily overfit to training data, especially in scenarios where data is limited.

These challenges highlighted the need for a new type of architecture, one that could offer improved interpretability and efficiency without compromising on performance.

#### The Kolmogorov-Arnold Theorem: The Foundation of KANs

The Kolmogorov-Arnold theorem serves as the theoretical foundation for the KAN architecture. In its simplest form, the theorem states that any continuous, multivariable function can be decomposed into a finite sum of continuous univariate functions, each applied to linear combinations of the input variables. This decomposition implies that a neural network can approximate any continuous function with a specific structure of interconnected functions.

#### Key Characteristics of KANs

1. **Learnable Activation Functions**: Unlike MLPs, where activation functions are fixed (e.g., sigmoid, ReLU), KANs employ learnable activation functions on edges, typically parameterized as splines. This flexibility allows each edge to dynamically adjust its contribution to the network's overall function.
   
2. **Spline-Based Architecture**: By using splines, which are piecewise polynomial functions, KANs can represent complex, non-linear relationships between inputs and outputs. Splines are advantageous because they can locally adjust activation functions without affecting other parts of the input space, leading to highly accurate modeling.

3. **Layered Structure**: The network consists of layered nodes connected via splines, allowing it to learn both compositional structures and univariate functions simultaneously. This layered approach helps KANs handle high-dimensional problems effectively.

#### Differences from Traditional Neural Networks

1. **No Linear Weights**: Traditional MLPs rely on linear weights between nodes, while KANs eliminate these weights entirely, relying instead on spline functions.

2. **Non-Linearity**: Non-linearities in KANs are embedded within the spline activation functions on edges rather than through separate activation layers.

3. **Compositional Structure**: KANs can learn both compositional structure (external degrees of freedom) and univariate functions (internal degrees of freedom), making them versatile in representing complex functions.

#### Practical Applications and Future Implications

The promise of KANs lies in their potential to revolutionize various domains of machine learning, from natural language processing to data fitting. Some key potential applications include:
- **Interpretability in Healthcare**: KANs can provide clearer insights into predictive modeling, improving trust in AI-driven diagnostic systems.
- **Reduced Computational Requirements**: Their parameter efficiency allows for faster training and deployment in resource-constrained environments.
- **Generative AI**: In large language models (LLMs), KANs offer significant improvements in interpretability, data efficiency, and modular knowledge representation.

Despite these potential advantages, challenges remain, particularly regarding training efficiency. Nevertheless, KANs represent a new era in neural network architecture, one that blends the strengths of existing models with novel approaches.
## The Genesis of KAN: Roots in the Kolmogorov-Arnold Theorem

The origins of KANs are rooted in the Kolmogorov-Arnold representation theorem, a result in mathematics that asserts any continuous, multivariate function can be decomposed into a sum of continuous univariate functions. This theorem, initially theoretical, posits that the complex relationships in a high-dimensional space can be effectively captured using combinations of simpler, one-dimensional functions.

This concept inspired researchers to develop a neural network architecture that could:
1. **Handle Complex Relationships**: Capture intricate relationships between inputs and outputs, particularly when the inputs are high-dimensional and non-linear.
   
2. **Be Modular and Interpretable**: By breaking down high-dimensional functions into univariate components, researchers anticipated that the networks would become easier to analyze and understand.

The Kolmogorov-Arnold theorem served as a cornerstone in designing a model that could address the interpretability and flexibility limitations of existing architectures.

### Defining Features of KANs

1. **Learnable Spline Functions**: KANs utilize splines as learnable functions on the edges of the network. Splines are piecewise polynomial functions that approximate non-linear relationships well. The flexibility to learn spline parameters directly on each edge allows the network to represent complex patterns.

2. **Summation Nodes**: The nodes in a KAN perform simple summation operations, aggregating the signals coming through spline functions. This operation differs from conventional nodes that apply activation functions to the summed signals.

3. **Hierarchical Organization**: Similar to other neural networks, KANs are structured hierarchically, enabling them to learn compositional structures across layers.

### How KANs Differ from MLPs and Other Networks

1. **Activation Functions on Edges**: Unlike MLPs, where fixed activation functions are placed after summing signals at nodes, KANs use learnable activation functions on edges, providing granular control over signal transformation.

2. **Spline-Based Learning**: The choice of spline functions as the primary mechanism for signal transformation differentiates KANs from traditional architectures that typically use simple activation functions like ReLU or sigmoid.

3. **Adaptability**: The modular design of KANs makes them adaptable to different data types and structures. Their ability to learn both compositional structure and univariate functions provides greater versatility.

### Current Applications and Research Trends

1. **Generative Models**: KANs are being explored for their potential to improve generative models like large language models (LLMs). Their interpretability and modularity can significantly enhance existing architectures by reducing parameter counts while improving efficiency.
   
2. **Scientific Discovery**: Their ability to solve complex mathematical problems, such as partial differential equations (PDEs), positions KANs as valuable tools in scientific computing and engineering.

3. **Healthcare**: The improved interpretability of KANs could revolutionize AI applications in healthcare, where understanding model decisions is crucial.

### Challenges and Limitations

Despite their promising potential, KANs are not without challenges. Key areas of concern include:
1. **Training Speed**: KANs often require more time to train due to their reliance on spline functions. Optimizing the training process remains an active research area.
   
2. **Stability**: The novel nature of KANs sometimes leads to unstable training, particularly in highly non-linear problems.

3. **Scalability**: Applying KANs to extremely large datasets or tasks may require further refinement in architecture design and training methodologies.
### Mathematical Aspects of Kolmogorov-Arnold Networks (KANs)

Kolmogorov-Arnold Networks (KANs) are inspired by the profound mathematical concepts underpinning the Kolmogorov-Arnold representation theorem. This section delves deep into the mathematical framework of KANs, exploring the principles that enable their unique functionality and the innovative approaches they bring to neural network architectures. We will discuss the formulation of KANs, the role of splines, and how these elements collectively support complex function approximation, providing a solid foundation for understanding their operation and capabilities.

#### Foundational Mathematical Concepts

1. **Kolmogorov-Arnold Representation Theorem**: Central to KANs is the theorem which states that any continuous function of several variables can be represented as a superposition of continuous functions of one variable. In practical terms, this theorem implies that complex multi-dimensional data relationships can effectively be broken down and learned using simpler, one-dimensional mappings.

2. **Function Decomposition**: At its core, a KAN decomposes high-dimensional functions into simpler components through the network. This decomposition is achieved by configuring the network to learn and represent these components as separate univariate functions, structured along the network's edges.

#### Spline Theory in KANs

1. **Usage of Splines**: In KANs, splines are employed as the functional units within the network’s architecture. Splines, especially cubic splines, are known for their flexibility and ability to provide a good approximation to complex functions. They are piecewise polynomials, defined within specific intervals, making them highly effective at local adaptations.

2. **Properties of Splines**: Splines maintain continuity and differentiability, which are critical for ensuring smooth transitions in the learned functions. The flexibility of splines allows them to fit a wide variety of data shapes, making them ideal for the edge functions in KANs.

3. **Optimization of Spline Parameters**: The training of a KAN involves adjusting the parameters of the spline functions to minimize error in the output. This process often uses gradient descent or other optimization techniques common in machine learning, tailored to handle the peculiarities of spline functions.

#### Compositional Structure Learning

1. **Learning the Compositional Structure**: KANs learn not only the individual univariate functions but also how these functions combine to form the overall output. This dual learning mechanism—learning both the functions and their compositions—is crucial for effectively modeling complex relationships inherent in real-world data.

2. **External Degrees of Freedom**: These refer to the ways in which the outputs of individual spline functions are combined. KANs optimize these combinations through their layered architecture, ensuring that the network learns the most effective way to integrate these functions.

#### Univariate Function Optimization

1. **Internal Degrees of Freedom**: Each spline function provides a degree of freedom that allows the KAN to fine-tune its response to variations in specific dimensions of the input data. By optimizing these internal degrees of freedom, KANs can achieve high levels of accuracy in function approximation.

2. **Local Adjustability**: The inherent property of splines to be adjusted locally without affecting other parts of the function makes them particularly suitable for tasks where different regions of the input space exhibit varied behaviors.

#### Integration of Mathematical Theories

The integration of these mathematical theories into KANs results in a powerful architecture capable of tackling complex, high-dimensional problems efficiently. This integration allows KANs to:

1. **Represent Complex Functions**: By combining multiple simple functions, KANs can represent highly complex and non-linear relationships, often with fewer parameters than required by traditional neural networks.

2. **Enhance Interpretability**: The mathematical structure of KANs, particularly the clear delineation of functions along the edges, enhances the interpretability of the model. This is a significant advantage in applications where understanding the model's decision process is crucial.

3. **Achieve High Precision**: The ability of KANs to fine-tune local behaviors of the model leads to high precision in predictions, particularly useful in scientific and engineering applications where accuracy is paramount.
### Upsides of Kolmogorov-Arnold Networks (KANs)

Kolmogorov-Arnold Networks (KANs) bring a host of compelling advantages that could reshape the landscape of machine learning. Their unique architecture, rooted in both theoretical and practical advancements, offers significant benefits that address some of the most pressing challenges faced by traditional neural networks. This section explores the major upsides of KANs, including improved accuracy, enhanced interpretability, and reduced computational requirements, among others.

#### 1. Improved Accuracy and Performance

1. **Superior Function Approximation**: KANs excel at accurately approximating complex functions, particularly in high-dimensional problems. By learning both the compositional structure and univariate functions, KANs deliver precise results across a wide range of tasks.

2. **Adaptive Spline-Based Learning**: The use of splines as activation functions provides KANs with local adjustability, allowing them to fine-tune their response in specific regions of the input space. This leads to a better fit and more accurate modeling compared to conventional activation functions.

3. **State-of-the-Art Performance**: Research indicates that KANs can outperform traditional MLPs and other architectures in tasks like data fitting, solving partial differential equations, and classification problems, especially when the models are carefully trained.

#### 2. Enhanced Interpretability

1. **Transparent Functional Blocks**: Each edge in a KAN is associated with a learnable spline function, which can be analyzed and visualized to understand how the network processes inputs. This makes KANs inherently more interpretable than traditional neural networks.

2. **Uncovering Relationships**: The layered structure of KANs and their spline-based activation functions can reveal the relationships between different input features and the model's decision-making process, making them valuable in domains where interpretability is crucial, such as healthcare and finance.

3. **Symbolic Regression Capabilities**: KANs can be used for symbolic regression, providing insight into the mathematical relationships between inputs and outputs. This feature is particularly useful in scientific research and engineering applications.

#### 3. Reduced Computational Requirements

1. **Parameter Efficiency**: KANs require fewer parameters to represent complex functions compared to traditional neural networks. This efficiency reduces the model's memory footprint and computational demands.

2. **Optimized Training**: By optimizing the architecture and training process, KANs can achieve comparable or superior results to larger networks, reducing the need for extensive computational resources during both training and inference.

3. **Fewer Data Requirements**: Their ability to effectively capture relationships allows KANs to generalize better with less training data, making them suitable for scenarios where data is limited or expensive to obtain.

#### 4. Generalization and Robustness

1. **Overfitting Resistance**: KANs are designed to focus on learning both compositional and univariate structures, which helps them generalize well to new data while avoiding overfitting to training data.

2. **Adaptability to Noise**: Their spline-based activation functions can adapt to noise and outliers in the data, making KANs more robust than traditional neural networks that use global activation functions.

#### 5. Potential for Multimodal Learning

1. **Combining Data Modalities**: KANs' ability to learn compositional structures makes them well-suited for combining and processing data from multiple modalities, such as text, images, and audio. This feature makes them particularly promising for tasks like multimodal classification and understanding.

2. **Knowledge Representation**: The modular structure of KANs could enable them to integrate knowledge representation more effectively, leading to better reasoning capabilities in large-scale AI applications.

### Downsides of Kolmogorov-Arnold Networks (KANs)

While Kolmogorov-Arnold Networks (KANs) introduce a promising neural network architecture with significant advantages, they also come with a set of challenges and limitations that should be addressed before widespread adoption. Understanding these downsides is essential for researchers and practitioners to optimize KANs or consider alternatives when necessary. Here, we explore the key drawbacks of KANs, including training complexity, computational inefficiency, and architectural challenges.

#### 1. Slow Training Speed

1. **High Computational Cost**: KANs rely on learnable spline functions, which significantly increase the computation required during training. The complexity of optimizing spline parameters results in training speeds often up to ten times slower than traditional MLPs.

2. **Gradient Optimization Challenges**: Training KANs requires specialized optimization methods due to the non-linear nature of spline functions. These challenges lead to longer convergence times, limiting KANs' practicality in environments where training speed is critical.

3. **Resource-Intensive Fine-Tuning**: Fine-tuning spline parameters can demand more computational resources compared to adjusting traditional activation functions. This requirement further adds to the overall training cost.

#### 2. Architectural Complexity

1. **Spline Parameterization**: While splines offer high flexibility, their parameterization can be tricky, requiring careful design and selection to ensure stable training. Poor parameterization can lead to suboptimal performance and increased training times.

2. **Lack of Standardization**: KANs lack standard architectural patterns, making it difficult to develop one-size-fits-all models. This limitation means practitioners often have to design and customize KANs for specific tasks, requiring specialized knowledge and time.

3. **Interdependencies Between Layers**: The layers in KANs have interdependencies that make the architecture more intricate to manage, often necessitating more sophisticated tools and expertise than standard MLPs.

#### 3. Stability and Scalability Issues

1. **Unstable Training**: Due to the high reliance on non-linear splines and the complexity of their learning process, training KANs can be unstable, particularly when applied to highly non-linear problems.

2. **Scalability Constraints**: Scaling KANs to handle very large datasets or tasks with high input dimensions may be challenging. As the number of input dimensions increases, the number of spline parameters required also rises exponentially, posing a scalability barrier.

3. **Memory Overhead**: Storing spline parameters for each edge can quickly lead to significant memory overhead, making KANs less feasible for memory-constrained environments or mobile applications.

#### 4. Interpretability Trade-offs

1. **Complex Interpretability**: While KANs are touted for their interpretability, the intricate nature of their architecture can sometimes obscure insights into the learned functions, particularly when models grow in size and complexity.

2. **Function Visualization Difficulties**: Visualizing learned spline functions across high-dimensional spaces is challenging, limiting the utility of interpretability in practical scenarios.

#### 5. Research Maturity

1. **Lack of Extensive Research**: The research into KANs is still relatively new, meaning that best practices and optimization strategies are still being developed. This lack of maturity makes it difficult for researchers to adopt and apply KANs confidently.

2. **Few Benchmark Comparisons**: Limited comparisons between KANs and other neural network architectures across different tasks and datasets make it difficult to establish clear guidelines for when KANs should be used.

```python
from kan.spline import B_batch
import torch
import matplotlib.pyplot as plt
import numpy as np

# consider a 1D example.
# Suppose we have grid in [-1,1] with G intervals, spline order k
G = 5
k = 3
grid = torch.linspace(-1,1,steps=G+1)[None,:]

# and we have sample range in [-1,1]
x = torch.linspace(-1,1,steps=1001)[None,:]

basis = B_batch(x, grid, k=k)

for i in range(G+k):
    plt.plot(x[0].detach().numpy(), basis[0,i,:].detach().numpy())

plt.legend(['B_{}(x)'.format(i) for i in np.arange(G+k)])
plt.xlabel('x')
plt.ylabel('B_i(x)')
```

```python
from kan import KAN

model = KAN(width=[1,1], grid=G, k=k)
# obtain coefficients c_i
model.act_fun[0].coef
assert(model.act_fun[0].coef[0].shape[0] == G+k)

# the model forward
model_output = model(x[0][:,None])

# spline output
spline_output = torch.einsum('i,ij->j',model.act_fun[0].coef[0], basis[0])[:,None]

torch.mean((model_output - spline_output)**2)
# residual output
residual_output = torch.nn.SiLU()(x[0][:,None])
scale_base = model.act_fun[0].scale_base
scale_sp = model.act_fun[0].scale_sp
torch.mean((model_output - (scale_base * residual_output + scale_sp * spline_output))**2)
```

```python
from kan import *
import numpy as np
import torch
import matplotlib.pyplot as plt


datasets = []

n_peak = 5
n_num_per_peak = 100
n_sample = n_peak * n_num_per_peak

x_grid = torch.linspace(-1,1,steps=n_sample)

x_centers = 2/n_peak * (np.arange(n_peak) - n_peak/2+0.5)

x_sample = torch.stack([torch.linspace(-1/n_peak,1/n_peak,steps=n_num_per_peak)+center for center in x_centers]).reshape(-1,)


y = 0.
for center in x_centers:
    y += torch.exp(-(x_grid-center)**2*300)

y_sample = 0.
for center in x_centers:
    y_sample += torch.exp(-(x_sample-center)**2*300)


plt.plot(x_grid.detach().numpy(), y.detach().numpy())
plt.scatter(x_sample.detach().numpy(), y_sample.detach().numpy())
```

```python
plt.subplots(1, 5, figsize=(15, 2))
plt.subplots_adjust(wspace=0, hspace=0)

for i in range(1,6):
    plt.subplot(1,5,i)
    group_id = i - 1
    plt.plot(x_grid.detach().numpy(), y.detach().numpy(), color='black', alpha=0.1)
    plt.scatter(x_sample[group_id*n_num_per_peak:(group_id+1)*n_num_per_peak].detach().numpy(), y_sample[group_id*n_num_per_peak:(group_id+1)*n_num_per_peak].detach().numpy(), color="black", s=2)
    plt.xlim(-1,1)
    plt.ylim(-1,2)
```

```python
ys = []

# setting bias_trainable=False, sp_trainable=False, sb_trainable=False is important.
# otherwise KAN will have random scaling and shift for samples in previous stages

model = KAN(width=[1,1], grid=200, k=3, noise_scale=0.1, bias_trainable=False, sp_trainable=False, sb_trainable=False)

for group_id in range(n_peak):
    dataset = {}
    dataset['train_input'] = x_sample[group_id*n_num_per_peak:(group_id+1)*n_num_per_peak][:,None]
    dataset['train_label'] = y_sample[group_id*n_num_per_peak:(group_id+1)*n_num_per_peak][:,None]
    dataset['test_input'] = x_sample[group_id*n_num_per_peak:(group_id+1)*n_num_per_peak][:,None]
    dataset['test_label'] = y_sample[group_id*n_num_per_peak:(group_id+1)*n_num_per_peak][:,None]
    model.train(dataset, opt = 'LBFGS', steps=100, update_grid=False);
    y_pred = model(x_grid[:,None])
    ys.append(y_pred.detach().numpy()[:,0])
```

```python
plt.subplots(1, 5, figsize=(15, 2))
plt.subplots_adjust(wspace=0, hspace=0)

for i in range(1,6):
    plt.subplot(1,5,i)
    group_id = i - 1
    plt.plot(x_grid.detach().numpy(), y.detach().numpy(), color='black', alpha=0.1)
    plt.plot(x_grid.detach().numpy(), ys[i-1], color='black')
    plt.xlim(-1,1)
    plt.ylim(-1,2)
```

# Scaling

```python
from kan import *

# initialize KAN with G=3
model = KAN(width=[2,1,1], grid=3, k=3)

data_sizes = np.array([100,300,1000,3000])
grids = np.array([5,10,20,50,100])

train_losses = np.zeros((data_sizes.shape[0], grids.shape[0]))
test_losses = np.zeros((data_sizes.shape[0], grids.shape[0]))
steps = 50
k = 3

for j in range(data_sizes.shape[0]):
    data_size = data_sizes[j]
    print(f'data_size={data_size}')
    # create dataset
    f = lambda x: torch.exp(torch.sin(torch.pi*x[:,[0]]) + x[:,[1]]**2)
    dataset = create_dataset(f, n_var=2, train_num=data_size)

    for i in range(grids.shape[0]):
        print(f'grid_size={grids[i]}')
        if i == 0:
            model = KAN(width=[2,1,1], grid=grids[i], k=k)
        if i != 0:
            model = KAN(width=[2,1,1], grid=grids[i], k=k).initialize_from_another_model(model, dataset['train_input'])
        results = model.train(dataset, opt="LBFGS", steps=steps, stop_grid_update_step = 30)
        train_losses[j][i] = results['train_loss'][-1]
        test_losses[j][i] = results['test_loss'][-1]
```

```python
for i in range(data_sizes.shape[0]):
    plt.plot(grids, train_losses[i,:], marker="o")
plt.xscale('log')
plt.yscale('log')
plt.plot(np.array([5,100]), 0.1*np.array([3,100])**(-4.), ls="--", color="black")
plt.legend([f'data={data_sizes[i]}' for i in range(data_sizes.shape[0])]+[r'$N^{-4}$'])
plt.ylabel('train RMSE')
plt.xlabel('grid size')
```

```python
for i in range(data_sizes.shape[0]):
    plt.plot(grids, test_losses[i,:], marker="o")
plt.xscale('log')
plt.yscale('log')
plt.plot(np.array([5,100]), 0.1*np.array([3,100])**(-4.), ls="--", color="black")
plt.legend([f'data={data_sizes[i]}' for i in range(data_sizes.shape[0])]+[r'$N^{-4}$'])
plt.ylabel('test RMSE')
plt.xlabel('grid size')
```

Fix model (grid) size, study data size scaling. No clear power law scaling. But we observe that: (1) increasing data size has no harm to performance. (2) powerful model (larger grid size) can benefit more from data size increase. Ideally one would want to increase data size and model size together so that their complexity always match.

```python
from kan import KAN
import torch
import copy


seed = 1

model = KAN(width=[6,1,1], grid=3, k=3, seed=seed)

# create dataset


def create_dataset(train_num=500, test_num=500):

    def generate_contrastive(x):
        # positive samples
        batch = x.shape[0]
        x[:,2] = torch.exp(torch.sin(torch.pi*x[:,0])+x[:,1]**2)
        x[:,3] = x[:,4]**3

        # negative samples
        def corrupt(tensor):
            y = copy.deepcopy(tensor)
            for i in range(y.shape[1]):
                y[:,i] = y[:,i][torch.randperm(y.shape[0])]
            return y

        x_cor = corrupt(x)
        x = torch.cat([x, x_cor], dim=0)
        y = torch.cat([torch.ones(batch,), torch.zeros(batch,)], dim=0)[:,None]
        return x, y

    x = torch.rand(train_num, 6) * 2 - 1
    x_train, y_train = generate_contrastive(x)

    x = torch.rand(test_num, 6) * 2 - 1
    x_test, y_test = generate_contrastive(x)

    dataset = {}
    dataset['train_input'] = x_train
    dataset['test_input'] = x_test
    dataset['train_label'] = y_train
    dataset['test_label'] = y_test
    return dataset

dataset = create_dataset()
```

```python
model(dataset['train_input'])
model.plot(beta=10)
```
